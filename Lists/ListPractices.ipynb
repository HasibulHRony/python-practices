```python
courses = ["History", "Math", "Physics", "CompSci"]
print(courses)
print(len(courses))
print(courses[0])
print(courses[3])
print(courses[0:4])
print(courses[0:2])
print(courses[1:3])
print(courses[-1])
print(courses[-3])
courses.append("Arts")
courses.insert(0, "Science")
courses_2 = ["Accounting", "Finance"]
courses.insert(2, courses_2)
print(courses)
courses.extend(courses_2)
print(courses)
courses.reverse()
print(courses)
```
