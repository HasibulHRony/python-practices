```python
#✅ Task 7: (Very Important for ML mindset)

#Generate 10 random numbers between 1–100 and store in a list.

#👉 Example output:

#[23, 67, 12, 89, 45, ...]


#✅ Task 8:

#From that list:

#Find max
#Find min
#Find average

#👉 This is EXACTLY how ML starts (data handling)


#Both the solution is given in the same code 



i = 1
numList = []
sum = 0
maxNum = 0
minNum = 100
while i<=10:
    num = random.randint(1, 100)
    sum = sum + num
    numList.append(num)
    if num > maxNum:
        maxNum = num
    
    if num < minNum:
        minNum = num
    i += 1

avg = sum / len(numList)
print("List: ", numList, "Sum: ", sum, "Min num: ", minNum, "Max num: ", maxNum, "Avg num: ", avg)

```
